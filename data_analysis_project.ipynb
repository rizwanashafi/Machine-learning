# **Data analysis and visualization**

```python
cancelled_perce= df['is_canceled'].value_counts(normalize = True)
cancelled_perce
```

```python
plt.figure(figsize = (10,8))
plt.title("resevation states count")
plt.bar(['not_cancelled', 'cancelled'], df['is_canceled'].value_counts(), edgecolor = 'black', linewidth= 3, width=0.5)
plt.show()
```

```python

plt.figure(figsize = (8,4))
axl = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')

legend_labels,_ = axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor = (1,1))
plt.title("Reservation status in different hotels", size = 20)
plt.xlabel("hotel")
plt.ylabel("number of reservations")
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'city Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby("reservation_status_date")[['adr']].mean()
city_hotel = city_hotel.groupby("reservation_status_date")[['adr']].mean()
```

```python
plt.figure(figsize = (10,8))
plt.title("Average Daily Rate in City and Resort Hotel", fontsize = 15)

plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'resort hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'city hotel')

plt.legend(fontsize = 15)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (15,8))

axl = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
legend_labels,_ = axl.get_legend_handles_labels()
# axl_legend(bbox_to_anchor=(1,1))
plt.title("reservation status per month", size = 20)
plt.xlabel("month")
plt.ylabel("number of reservation")
# plt_legend(["not_canceled", "canceled"])
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.title('adr per month', fontsize = 20)

# sns.barplot('month', 'adr', data = df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
canceled_data = df[df['is_canceled'] == 1]
total_adr_per_month = canceled_data.groupby('month')['adr'].sum().reset_index()
sns.barplot(x='month', y='adr', data=total_adr_per_month, palette='bright')

# plt.legend(fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (20,8))

canceled_data = df[df['is_canceled'] == 1]
top_10_country = canceled_data['country'].value_counts()[:10]
# sns.pie('top 10 country' , autopct = '%.2f', labels = top 10 country.index)
plt.pie(x=top_10_country.values, labels=top_10_country.index, autopct='%.2f')

plt.title('Top 10 Countries Distribution', fontsize=20)
plt.show()
```

```python
df = pd.read_csv("/content/hotel_booking.csv")

canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace = True)
canceled_df_adr.sort_values('reservation_status_date', inplace = True)

not_canceled_data = df[df["is_canceled"] == 0]
not_canceled_df_adr = not_canceled_data.groupby('reservation_status_date')['adr'].mean().reset_index()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,8))
plt.title('reservation status date')
plt.plot(not_canceled_df_adr['reservation_status_date'], not_canceled_df_adr['adr'],label = 'not_canceled')
plt.plot(not_canceled_df_adr['reservation_status_date'], not_canceled_df_adr['adr'],label = 'canceled')
plt.show()
```
