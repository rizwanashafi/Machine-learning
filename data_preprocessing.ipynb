```python
# visualize correlation using heat map
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,5))

correlation = df.corr()
sns.heatmap(correlation, cmap='YlOrRd', annot = True)
```
