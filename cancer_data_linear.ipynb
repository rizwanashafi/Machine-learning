```python
# Select the columns containing categorical data
categorical_columns = ['Diagnosis']
```

```python
from sklearn.preprocessing import LabelEncoder
# Perform label encoding for each categorical column
le = LabelEncoder()
for column in categorical_columns:
    df['Diagnosis'] = le.fit_transform(df['Diagnosis'])
```

```python
# smmary
df.describe()
```

```python
# Target variable
x = df.drop(["Fractal dimension (worst)", "Diagnosis"], axis = 1)
y = df["Diagnosis"]
```

```python
# spliting data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 102)
```

```python
# standard daviation
from sklearn.preprocessing import StandardScaler
Sc = StandardScaler()
x_train = Sc.fit_transform(x_train, x_test)
```

```python
# Model evaluation
from sklearn.linear_model import LinearRegression
ln_reg = LinearRegression()
ln_reg.fit(x_train, y_train)
```

```python
# predict
y_pred = ln_reg.predict(x_test)
y_pred
```

```python
# define actual values
results = np.column_stack((y_test, y_pred))

print("actuall values | predicted values")
print("-----------------------")
for actual, predicted in results:
  print(f"{actual:10.2f} | {predicted:8.2f}")
```

```python
residual = actual- y_pred.reshape(-1)
print(residual)
```

```python
import seaborn as sns
sns.distplot(residual, kde=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x=y_test, y=y_pred, color='blue', label='Actual Data points')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', label='Ideal Line')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual (Linear Regression)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
print("Linear Regression Model")
print("--"*30)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("mean squared error", mse)
print("root mean squared error", rmse)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(x_train, y_train)
# predt
y_pred_dtr = dtr.predict(x_test)
dtr= mean_squared_error(y_pred_dtr, y_test)
print('Decision Tree Regression : ',dtr)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

rf_regressor = RandomForestRegressor()
rf_regressor.fit(x_train,y_train)

#Predicting the SalePrices using test set
y_pred_rf = rf_regressor.predict(x_test)
RFr = mean_squared_error(y_pred_rf,y_test)
#Random Forest Regression Accuracy with test set
print('Random Forest Regression : ',RFr)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

rf_regressor = GradientBoostingRegressor()
rf_regressor.fit(x_train,y_train)

#Predicting the SalePrices using test set
y_pred_rf = rf_regressor.predict(x_test)

#Random Forest Regression Accuracy with test set
GBr = mean_squared_error(y_pred_rf,y_test)
print('Gradient Boosting Regression : ',GBr)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor()
knr.fit(x_train, y_train)

# predt
y_pred_knr = knr.predict(x_test)
knr= mean_squared_error(y_pred_knr, y_test)

print('K Neighbors Regressor : ',knr)
```
